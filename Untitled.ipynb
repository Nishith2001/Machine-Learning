```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
df = pd.read_csv("/content/diabetes.csv")
df
```

```python
column_names = ['Pregnancies', 'Glucose', 'BP', 'SkinThickness', 'Insulin','BMI','DBF','Age', 'label']
diabetes_data = pd.read_csv("diabetes.csv", header=None, names= column_names)
diabetes_data.head()
```

```python
diabetes_data= diabetes_data.iloc[1:]
diabetes_data.head()
```

```python
feature_columns = ['Pregnancies', 'Glucose', 'BP', 'SkinThickness', 'Insulin','BMI','DBF','Age'  ]
x= diabetes_data[feature_columns]
y= diabetes_data.label
```

```python
from sklearn.model_selection import train_test_split
train_x,train_y,test_x,test_y = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000)
```

```python
logreg.fit = (train_x,train_y)
```
